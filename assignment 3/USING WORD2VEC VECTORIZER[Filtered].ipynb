```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
from tabulate import tabulate
from collections import Counter
```

```python
df = pd.read_csv('news_dataset.csv')

# Extract the 'text' column as a list
dataset = df['text'].dropna().tolist()

print(dataset[:5])
```

```python
def convert_to_lowercase(text):
    return text.lower()
dataset = df['text'].dropna().tolist()

# Convert each text to lowercase
dataset = [convert_to_lowercase(text) for text in dataset]

# Print only the first 20 items to avoid overloading the console
for i, text in enumerate(dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
import re
import string

def remove_punctuation(text):
    if not isinstance(text, str):
        return text
    text = re.sub(r'[\n\r]+', ' ', text)
    punctuation_to_remove = string.punctuation.replace("'", "")
    text = re.sub(r"[{}]".format(re.escape(punctuation_to_remove)), " ", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

dataset = [remove_punctuation(text) for text in dataset]

for i, text in enumerate(dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
from nltk.corpus import stopwords

def remove_stopwords(text):
    if isinstance(text, str):
        return ' '.join(word for word in text.split() if word.lower() not in stop_words)
    return text
stop_words = set(stopwords.words('english'))
# Apply all transformations
dataset = [remove_stopwords(text) for text in dataset]

# Print the cleaned dataset
for i, text in enumerate(dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
def remove_suffixes(text):
    if isinstance(text, str):
        words = text.split()
        stripped_words = []
        for word in words:
            # Basic handling for common suffixes
            if word.endswith("ing") and len(word) > 4:
                stripped_words.append(word[:-3])
            elif word.endswith("es") and len(word) > 3:
                stripped_words.append(word[:-2])
            else:
                stripped_words.append(word)
        return ' '.join(stripped_words)
    return text

# Apply the new step AFTER the previous processing
dataset = [remove_suffixes(text) for text in dataset]

# Print the final output
for i, text in enumerate(dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
import re
dataset = [text for text in dataset if not re.search(r'\d', text)]

for i, text in enumerate(dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
!pip install symspellpy
from symspellpy import SymSpell, Verbosity

# Create and load SymSpell
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
sym_spell.load_dictionary("frequency_dictionary_en_82_765.txt", term_index=0, count_index=1)

def autocorrect_text(text):
    if not isinstance(text, str):
        return text
    corrected_words = []
    for word in text.split():
        suggestions = sym_spell.lookup(word, Verbosity.CLOSEST, max_edit_distance=2)
        corrected_word = suggestions[0].term if suggestions else word
        corrected_words.append(corrected_word)
    return ' '.join(corrected_words)

# Apply correction to all dataset items
corrected_dataset = [autocorrect_text(text) for text in dataset]

# Show first 20 results
for i, text in enumerate(corrected_dataset[:20], start=1):
    print(f"{i}: {text}")
```

```python
tokenized_dataset = [doc.split() for doc in dataset]
word2vec_model = Word2Vec(sentences=tokenized_dataset, vector_size=100,
window=5, min_count=1, workers=4)
```

```python
import numpy as np

embedding_dim = word2vec_model.vector_size  # e.g., 300

X = []
for doc in dataset:
    vectors = [word2vec_model.wv[word] for word in doc.split() if word in word2vec_model.wv]
    if vectors:
        doc_vector = np.mean(vectors, axis=0)
    else:
        doc_vector = np.zeros(embedding_dim)  # or handle differently (skip, nan, etc.)
    X.append(doc_vector)

X = np.array(X)
```

```python
k = 2  # number of clusters
km = KMeans(n_clusters=k, random_state=42)
km.fit(X)

# Predict cluster for each document
y_pred = km.predict(X)

# Prepare and print only first 20 results
table_data = [["Document", "Predicted Cluster"]]
table_data.extend([[doc, cluster] for doc, cluster in zip(dataset[:20], y_pred[:20])])

print(tabulate(table_data, headers="firstrow"))
```

```python
# Calculate purity
total_samples = len(y_pred)
cluster_label_counts = [Counter(y_pred)]
purity = sum(max(cluster.values()) for cluster in cluster_label_counts) / total_samples
print("Purity:", purity)
```

```python
from gensim.models.coherencemodel import CoherenceModel

k = 2
clusters = y_pred

# Create a list of tokenized documents (list of words)
tokenized_docs = [doc.split() for doc in dataset]

# For each cluster, aggregate words
cluster_words = []
for cluster_id in range(k):
    # Combine all words from docs in this cluster
    words = []
    for i, c in enumerate(clusters):
        if c == cluster_id:
            words.extend(tokenized_docs[i])
    # Get top N words
    top_words = [word for word, freq in Counter(words).most_common(20)]
    cluster_words.append(top_words)

# Now compute coherence score for each cluster's top words
# We need a gensim Dictionary and corpus for the whole dataset
from gensim.corpora.dictionary import Dictionary

dictionary = Dictionary(tokenized_docs)
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

coherence_scores = []
for top_words in cluster_words:
    cm = CoherenceModel(
        topics=[top_words], 
        texts=tokenized_docs, 
        dictionary=dictionary, 
        corpus=corpus, 
        coherence='c_v'
    )
    score = cm.get_coherence()
    coherence_scores.append(score)

for idx, top_words in enumerate(cluster_words):
    cm = CoherenceModel(
        topics=[top_words],
        texts=tokenized_docs,
        dictionary=dictionary,
        corpus=corpus,
        coherence='c_v'
    )
    score = cm.get_coherence()
    print(f"Cluster {idx} top words: {top_words}")
    print(f"Cluster {idx} coherence score: {score:.4f}\n")
```

Cluster 1 is slightly more coherent (0.404) than Cluster 0 (0.376), but both scores are low, likely due to common words dominating. Better preprocessing is needed for clearer topics.
